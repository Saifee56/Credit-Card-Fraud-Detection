```python
import pandas as pd 
import numpy as np 
import random 
```

```python
df=pd.read_csv(r"C:\Users\my pc\Desktop\Credit_card_fraud\Credit-Card-Fraud-Detection\CardData_SAIDUR_V.csv")
df
```

```python
def map_gender(title):
    if title.startswith('Mr'):
        return 'M'
    elif title.startswith('Mrs') or title.startswith('Miss'):
        return 'F'
    else:
        return None  

df['gender'] = df['NAME'].apply(map_gender)

print(df)
```

```python
df['gender'].value_counts()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
def extract_first_name(address):
    return address.split("  ")[0]

# Apply the function to the address column
df['merchant'] = df['CARD_ACC_NAME_ADDRESS'].apply(extract_first_name)

# Print the updated dataset
print(df)
```

```python
categories = [
    'personal_care',
    'health_fitness',
    'misc_pos',
    'travel',
    'kids_pets',
    'shopping_pos',
    'food_dining',
    'home'
]

# Generate 'category' column with random values
df['category'] = random.choices(categories, k=len(df))

# Print the updated dataset
print(df)
```

```python
jobs = [
    'Engineer',
    'Teacher',
    'Doctor',
    'Artist',
    'Lawyer',
    'Chef',
    'Writer',
    'Nurse'
]

# Generate random values for 'job', 'unix_time', 'merch_lat', 'merch_long', and 'transaction_amount' columns
df['job'] = np.random.choice(jobs, len(df))
df['unix_time'] = np.random.randint(1325376018, 1371816817, len(df))  # Random Unix time between Jan 1, 2012 and June 21, 2020
df['merch_lat'] = np.random.uniform(300, 500, len(df))  # Random latitude between 30 and 50
df['merch_long'] = np.random.uniform(-120, -70, len(df))  # Random longitude between -120 and -70
df['transaction_amount'] = np.random.uniform(10000, 50000, len(df))  # Random transaction amount between 1 and 1000

# Print the updated dataset
print(df)
```
